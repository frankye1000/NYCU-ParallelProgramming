```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_array = cv2.imread("./Image_data/football.jpg")
```

```python
row,column,dim = img_array.shape
print(row,column,dim)
```

```python
with open('./Image_data/football.txt.membership') as f:
    contents = [line.strip("\n") for line in f.readlines()]

contents[:10]
```

```python
colordict = {'0': [66, 153, 244],
             '1': [234, 67, 53],
             '2': [251, 188, 5],
             '3': [52, 168, 83],
             '4': [127,255,212],
             
             
             
             
             
             }
```

```python
len(contents)
```

```python
img_array_group = np.zeros((row, column, dim))
```

```python
i = 0
for r in range(row):
    for c in range(column):
        group_index = contents[i].split(' ')[1]
        img_array_group[r][c] = colordict[group_index]
        i+=1
```

```python
plt.figure()
f, axarr = plt.subplots(1, 2,figsize=(10,8)) 
axarr[0].imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
axarr[1].imshow(img_array_group.astype(np.uint8))                                 # 在圖表中繪製圖片
plt.show()    
```
